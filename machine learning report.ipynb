## machine learning introduction


Defination: Machine learning (ML) is the study of computer algorithms that improve automatically through experience.

Techniques:
1. Unsupervised learning(Group and interpret data based only on input data)
   - Clustering -- Exploratory data analysis to find patterns or groupings.



2. Supervised learning(Develop predictive model based on both input and output data
   - Classifacation -- Used for discrete responses prediction. Classifacation models put input data into categories.
   - Regression -- Used for continues reponses prediction, like algorithmic trading.


Difficulty:
1. Real-world data can not be processed direcly
2. Preprocessing methods vary a lot
3. Fitting the dataset with best model is challenging.

## Machine learning example: Classify Physical Activities

1. Load the data
2. Preprocess the data
   - Look for outliers 
   - Check the missing value
   - Apply filter to the data
   - Divide the data into the test and training sets
3. Feature Engineering
   - create a new table that includes the selected features that can help the algorithm distinguish between different patterns.
   - Use various techniques to find feature withn various datatype.
4. Build and Train the model
   - Decision Treee
   - K-nearest neighbors(KNN)
   - multiclass support vector machine(SVM)
5. Improve the model
   - Simplify by correlation matrix, Principal component analysis, and sequential feature reduction
   - Add complexity by using model combination and adding more data sourses
   - Dimensionality Reduction Method
      - Principal component analysis(PCA): Generate new variables via linear combinations of the original variables.
      - Factor analysis
      - Nonnegative matrix factorization

## Supervised Learning


###Classification Algorithms

Logistic Regression
   - When data can be separated by a single linear boundary
   - baseline for classificaiton

k Nearest Neighbor(kNN)
   - categorize object based on the classes of nearest neighbors
   - distance metrics:
      - Euclidean
      - city block
      - cosine
      - Chebychev


Support Vector Machine
   - find the linear boundry(hyperplane) which that seperate data with the largest margin between two classes.
   - loss function is used to penalize wrong points
   - can be use for high dimensional, nonlinear data
   

Neural Network
   - Trained by iteratively modifying the strengths of the connections

Naive Bayes
   - classifies new data based on the highest probability of its belonging to a particular class

Discriminant Analysis
   - Finding linear combinations of features

Decision Tree
   - branching condition where trained weight is compared to the value

Bagged and boosted decision tree
   - when predictors are categorical

### Regression Algorithms


Linear Regression

SVM Regression
   - be able to predict a continuous response. 

Regresion Tree
